# Probability of new disease cases using Poisson Distribution

healthcare scenario where a hospital wants to estimate the incidence of a rare disease, such as a particular type of cancer, in their region. The goal is to predict how many new cases will be reported in a month based on historical data.

### Dataset 

```python
# Import necessary libraries
import pandas as pd

# Create a DataFrame with the historical data of new disease cases
data = {
    "Month": ["January", "February", "March", "April", "May", "June", 
              "July", "August", "September", "October", "November", "December"],
    "New Cases": [3, 2, 4, 5, 3, 4, 3, 2, 4, 3, 5, 2]
}

# Convert the data into a DataFrame
df = pd.DataFrame(data)
```

### Estimating probable cases

```python
# Import necessary libraries
import math
import matplotlib.pyplot as plt

# Mean rate (lambda) from the data
lambda_value = 3.33

# Function to calculate Poisson probability
def poisson_probability(k, lambda_value):
    return (math.exp(-lambda_value) * lambda_value**k) / math.factorial(k)

# Calculating probabilities for k = 0 to 5
k_values = range(6)
probabilities = [poisson_probability(k, lambda_value) for k in k_values]

# Plotting the Poisson distribution
plt.figure(figsize=(10, 6))
plt.bar(k_values, probabilities, color='Teal', alpha=0.7)
plt.xlabel('Number of New Cases (k)')
plt.ylabel('Probability P(X=k)')
plt.title('Poisson Distribution of New Disease Cases')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
